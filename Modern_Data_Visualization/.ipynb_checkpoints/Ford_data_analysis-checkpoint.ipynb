## Investigation Overview

 The aim of this presentation is to extract insights that will help the company to better exploit the resources and prioritize its segmentations and users by distinguishing the dominant behavior of different users.

## Dataset Overview

 The dataset is extracted from the publically-published FORD-GOBIKE dataset. This Analysis focuses on the first 6 months of the year 2018.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Frod_mod.csv')
df.start_time = pd.to_datetime(df.start_time)
df.end_time = pd.to_datetime(df.end_time)
plt.rcParams['font.size'] = 14
```

# How much distance are the bikes often used for?

__The upcoming graph shows the distribution of the distances traveled by the members. and the following could be concluded:__
1. Most renters use the bikes for relatively __short distances (< 2km).__
2. There are very few bikers who are traveling long distances, this shows that the bikes, overall, are just __used for minor purposes,__ probably just commuting to work or school.<br>
__So, companies shouldn't focus  their production lines on highly sophisticated bikes, rather, medium and adequate products.__

```python
plt.figure(figsize=(9,6))
df = df[df.distance < 10000]
g2 = sns.distplot(df.distance, kde=False, color = "#ED553B", hist_kws={'alpha':1}, bins=40);
g2.set(xscale='linear', title='Distance Traveled With Bikes');
plt.xlabel("Distance (m)")
plt.ylabel("Number of Members");
```

## Which category of members is the most distinguished?

__Another interesting information that the data suggests also is:__
1. Male Members are much more than Females and other gender Members.
2. Dominant year of birth $\approx$ 1981 -> 1995, which defines an age range of 23 -> 31. Most of who are graduates and employees.
3. The Least remarked users are thoes older than 31 year and belong other genders.

```python
grp = df.groupby(['member_birth_year', 'member_gender'])
grp_size = grp.distance.size().reset_index()
grp_size = grp_size.pivot(index='member_birth_year', columns='member_gender', values='distance')
grp_size.fillna(0, inplace=True)
grp_size.index = grp_size.index.astype(int)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=grp_size.loc[1970.0:,:], cbar_kws={'label':"Number of Members"}, cmap="viridis_r");
plt.yticks(ticks=[0, 5, 10, 15, 20, 25, 30],labels=[1970, 1975, 1980, 1985, 1990, 1995, 2000]);
plt.xlabel("Members Gender")
plt.ylabel("Birth Year")
plt.title("Number of Members vs. Year of Birth");
```

### Deeper into the Gender vs. Age Relation:
* A more tangible relation could be detected in the line plot graph, which __emphasizes that the age range observed previously is a common range for all genders__ (Males, Female, and Other). which provide a clear segmentation target for the company to concentrate on this age category (23 to 31 year old).

```python
plt.figure(figsize=(9,6))
sns.lineplot(data=grp_size, dashes=False, palette='Set1',hue_order=['Male', 'Female', "Other"])
plt.xlabel("Year of Birth")
plt.title("Number of Members per Year of Birth")
plt.ylabel("Number of members")
plt.legend(bbox_to_anchor=(0.25,1), title="Gender");
```

## How much time does different users usually use the bike for?
* It looks like the __customers are renting bikes more than subscribers__, which indicates that the __pay-as-you-go stategy is more preferable than the subsciption packages__.
* Secondly, In the customers, Females are spending more time which contradictes The Previous observation, and the only explanation is that __females are a bit slower than other genders__ when driving a bike.

```python
plt.style.use('default')
sns.barplot(y='duration_sec',x='user_type', data=df, hue='member_gender', palette="Set2", saturation=1)
plt.ylabel("Duaration (s)")
plt.xlabel("User Type")
plt.title("Average Time Per User Type & Gender", pad=10)
plt.legend(bbox_to_anchor=(1,1), title="Gender");
```

## Does the usage varies over time for the customers?
* Apparently, There was a strong relation between the __duration, distance and the user type. Customer-type users are the dominant users__; they have the highest duration and distances in all categories.
* Even __when viewing each month seperately, the customer-type users are still the highest__ in the distance traveled.

 __This should provide a hint for the company to elevate the offers in order to attract more subscribers, given that this is the company's vision.__

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='month', y='distance', hue='user_type', data=df, palette='Set2', saturation=1)
plt.xlabel("Month")
plt.ylabel("Distance (meters)")
plt.legend(bbox_to_anchor=(1,1), title="User Type")
plt.title("Distance Traveled by Each User Type Per Month", pad=20);
```

# Conclusion <br>

* Most members are Males and age from 23 to 31 years old, therefore, the company should consider these categories the most while providing the service in the future.<br>
* Customers are the dominant members in the bikes' usage. A proper strategy should be initiated in order to account for this section of users.
* Females are the slowest bikers, a more efficient bikes should be provided t the females in order to overcome this gap in distance vs. time between the genders.
